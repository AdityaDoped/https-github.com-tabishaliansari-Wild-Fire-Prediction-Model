```python
# Step 1: Group by 'date' and 'is_fire', and calculate the mean for all other columns
grouped_df = df.groupby(["acq_date", "is_fire"]).mean().reset_index()

# Step 2: Sort the resulting DataFrame by 'date' and 'is_fire' for better readability
grouped_df = grouped_df.sort_values(by=["acq_date"])

# Step 3: Verify the length of the resulting DataFrame
print(f"Number of rows in the grouped DataFrame: {len(grouped_df)}")

# Step 4: Display the resulting DataFrame
print(grouped_df)
```
